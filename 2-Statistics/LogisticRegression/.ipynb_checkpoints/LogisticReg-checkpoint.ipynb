# Import

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Sigmoid Function

### Sigmoid

```python
def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a
```

```python
x = np.arange(-3., 3., 0.2)
sig = sigmoid(x)
```

```python
plt.figure(figsize=(12,8))

plt.plot(x,sig, label='sigmoid')
plt.plot(x,x, label='input')

plt.title("Sigmoid Function")
plt.legend()
plt.show()
```

### Logit

```python
def logit(x):
    a = []
    for item in x:
        a.append(math.log(item/(1-item)))
    return a
```

```python
plt.figure(figsize=(12,8))

plt.plot(x,sig, label='sigmoid')
plt.plot(x,logit(sig), label='logit tranform')

plt.axvline(0, color='red', alpha=0.1)
plt.axhline(0.5, color='green', alpha=0.1)

plt.title("Sigmoid - Logit Function")
plt.legend()
plt.show()
```

# Implementation in Sk-learn

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

data = load_breast_cancer()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['data'], data['target'], test_size = 0.25)
```

### L1-Regularization

```python
lr = LogisticRegression(penalty='l1')
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
print(accuracy_score(y_pred, y_test))
print(f1_score(y_pred, y_test))
```

### L2-Regularization

```python
lr = LogisticRegression(penalty='l2', solver='lbfgs')
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
print(accuracy_score(y_pred, y_test))
print(f1_score(y_pred, y_test))
```

We can now illustrate the impact of the tolerance factor C. The larger C is, the less restrictive is the regularization.

```python
from sklearn.model_selection import GridSearchCV

parameters = {'C':[0.1, 0.5, 1, 2, 5, 10, 100]}
lr = LogisticRegression(penalty='l2', max_iter = 5000, solver='lbfgs')

clf = GridSearchCV(lr, parameters, cv=5)
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

C=10 appears to be the best model. 

```python
y_pred = clf.predict(X_test)
print(accuracy_score(y_pred, y_test))
print(f1_score(y_pred, y_test))
```
